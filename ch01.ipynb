## The lambdify() function

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, sin, diff
from sympy.utilities.lambdify import lambdify

a, b, t = symbols('a, b, t', real=True)
f = a*sin(b*t)
deriv = diff(f,t)  # evaluate symbolic derivarive of f
print('v = dy/dt = ', deriv)
```

```python
tims = np.linspace(0, 6, 50)
y = lambdify(t, f.subs({a:1, b:2}))
plt.subplot(1, 2, 1)  # first subplot, positions y(t)
plt.plot(tims, y(tims))
plt.xlabel('Time [s]')
plt.ylabel('y(t)')

u = deriv.subs({a:1, b:2})
v = lambdify(t, u)

plt.subplot(1, 2, 2)
plt.plot(tims, v(tims))
plt.xlabel('Time [s]')
plt.ylabel('Speed v(t)=dy/dt')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 13.Evaluating of product series for cosx
import numpy as np


def f(x, N):
    product = 1
    for k in range(1, N + 1):
        term = 1 - (4 * x ** 2) / (np.pi ** 2 * (2 * k - 1) ** 2)
        product *= term
    return product

# cos(1)
true_value = np.cos(1)

# N
N_values = [10, 50, 100]
for N in N_values:
    result = f(1, N)
    print(f"N = {N} : {result}")
    print(f" cos(1) : {np.abs(result - true_value)}")
    print("-" * 30)
```

```python
# 14. Evaluation of the sum of infinite series
from sympy import symbols, Sum, factorial, sin, cos, exp, log, pprint

# 
x, n, k = symbols('x n k')

# 
series1 = Sum((-1)**n * x**(2*n + 1) / factorial(2*n + 1), (n, 0, float('inf')))
result1 = series1.doit()

# 
series2 = Sum(x**(2*n) / factorial(2*n), (n, 0, float('inf')))
result2 = series2.doit()

# 
series3 = Sum((-1)**n * x**(2*n) / factorial(2*n), (n, 0, float('inf')))
result3 = series3.doit()

# 
series4 = Sum(x**k / factorial(k), (k, 0, float('inf')))
result4 = series4.doit()

# 
series5 = Sum(x**k / k, (k, 1, float('inf')))
result5 = series5.doit()

# 
print(":")
pprint(result1)
print("\n:")
pprint(result2)
print("\n:")
pprint(result3)
print("\n:")
pprint(result4)
print("\n:")
pprint(result5)
```

```python
# 17. The Bessel fucntions
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# 
x, n, p = sp.symbols('x n p')
# 
gamma = sp.gamma

# 
bessel_series = sp.Sum(((-1)**n / (gamma(n + 1) * gamma(n + 1 + p))) * (x / 2)**(2 * n + p), (n, 0, sp.oo))

# 
def bessel_function(p_val, x_val):
    result = bessel_series.subs({p: p_val, x: x_val}).doit()
    return result

#  x 
x_vals = np.linspace(-20, 20, 300)

#  p = 0  p = 2 
y_vals_0 = [float(bessel_function(0, x)) for x in x_vals]
y_vals_2 = [float(bessel_function(2, x)) for x in x_vals]

# 
plt.plot(x_vals, y_vals_0, label=r'$J_0(x)$')
plt.plot(x_vals, y_vals_2, label=r'$J_2(x)$')
plt.xlabel('x')
plt.ylabel('$J_p(x)$')
plt.title('Bessel Functions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 18. Plots of the sum of terms in a Fourier series
import numpy as np
import matplotlib.pyplot as plt

#  L
L = 5

#  x 
x = np.linspace(0, 10, 1000)

# 
def fourier_series_sum(N, x):
    sum_value = 0
    for n in range(1, N + 1):
        sum_value += ((-1)**(n - 1) / n) * np.sin((n * np.pi / L) * x)
    return (50 / np.pi) * sum_value - 25

#  N = 5, 10, 20, 30 
N_values = [5, 10, 20, 30, 100]
for N in N_values:
    y = fourier_series_sum(N, x)
    plt.plot(x, y, label=f'N = {N}')

# 
plt.xlabel('x')
plt.ylabel('T(x)')
plt.title('Fourier Series of Sawtooth Wave')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 19. Plots of the sum of terms in a Taylor series
import numpy as np
import matplotlib.pyplot as plt
import math


# 
def taylor_series_sum(N, x):
    sum_value = 0
    for n in range(N):
        term = ((-1) ** n / math.factorial(2 * n+1)) * (x ** (2 * n+1))
        sum_value += term
    return sum_value


# x
x = np.linspace(0, 8, 1000)

# N = 5, 6, 8, 10
N_values = [5, 6, 8, 10, 40]
for N in N_values:
    y = [taylor_series_sum(N, xi) for xi in x]
    plt.plot(x, y, label=f'N = {N}')

# 
plt.xlabel('x')
plt.ylabel('T(x)')
plt.title('Taylor Series of cos(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 21. The Pauli matrics
import numpy as np

# 
sigma_x = np.array([[0, 1], [1, 0]])
sigma_y = np.array([[0, -1j], [1j, 0]])
sigma_z = np.array([[1, 0], [0, -1]])
identity_matrix = np.array([[1, 0], [0, 1]])

# sigma_x^2 = I
result1 = np.dot(sigma_x, sigma_x)
assert np.allclose(result1, identity_matrix)

# sigma_y^2 = I
result2 = np.dot(sigma_y, sigma_y)
assert np.allclose(result2, identity_matrix)

# sigma_z^2 = I
result3 = np.dot(sigma_z, sigma_z)
assert np.allclose(result3, identity_matrix)

#  -i * sigma_x * sigma_y * sigma_z = I
result4 = -1j * np.dot(np.dot(sigma_x, sigma_y), sigma_z)
assert np.allclose(result4, identity_matrix)

# sigma_x * sigma_y = i * sigma_z
result5 = np.dot(sigma_x, sigma_y)
assert np.allclose(result5, 1j * sigma_z)

# sigma_x * sigma_y + sigma_y * sigma_x = 0
result6 = np.dot(sigma_x, sigma_y) + np.dot(sigma_y, sigma_x)
assert np.allclose(result6, np.zeros((2, 2)))

# sigma_x * sigma_y - sigma_y * sigma_x = 2i * sigma_z
result7 = np.dot(sigma_x, sigma_y) - np.dot(sigma_y, sigma_x)
assert np.allclose(result7, 2j * sigma_z)

print("")
```

```python
from sympy import symbols, simplify, I

# 
R, L, C, omega = symbols('R L C omega')
Z_R = R
Z_L = I * omega * L
Z_C = -I / (omega * C)

# Z
Z = Z_R + Z_L + Z_C

# Z
Z_simplified = simplify(Z)

# (a) Z
real_part = simplify(Z_simplified.coeff(I, 0))
imag_part = simplify(Z_simplified.coeff(I, 1))

print("Z", real_part)
print("Z", imag_part)

# (b) Z
R_value = 5
L_value = 1e-3
C_value = 1e-3
omega_value = 10

Z_value = Z_simplified.subs({R: R_value, L: L_value, C: C_value, omega: omega_value})

print("Z", Z_value)
```

```python
# 26. Finding maxima, minima and inflection points
from sympy import symbols, diff, solve, Eq, simplify

# 
x = symbols('x', positive=True)

# 
f = x**(-6) - x**(-12)

# 
f_prime = diff(f, x)

# 
f_double_prime = diff(f, x, 2)

# 0
critical_points = solve(Eq(f_prime, 0), x)

# 0
inflection_points = solve(Eq(f_double_prime, 0), x)

# /
max_min_points = []
for cp in critical_points:
    # 
    result = f_double_prime.subs(x, cp)
    if result > 0:
        max_min_points.append((cp, "min"))
    elif result < 0:
        max_min_points.append((cp, "max"))

print(":")
print(f_prime)
print(":")
print(f_double_prime)
print("/:")
for point in max_min_points:
    print(f"x = {point[0]}, : {point[1]}")
print(":")
for ip in inflection_points:
    print(ip)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  J/K
k_B = 1.380649e-23
# 
eV_to_J = 1.602176634e-19

# Bose - Einstein
def bose_einstein_distribution(E, T):
    E_J = E * eV_to_J
    denominator = np.exp(E_J / (k_B * T)) - 1
    # 
    result = np.where(denominator != 0, 1 / denominator, np.nan)
    return result

#  eV
E_values = np.linspace(0, 30, 1000)

#  K
T_values = [2000, 4000, 6000]

# 
for T in T_values:
    n_values = bose_einstein_distribution(E_values, T)
    plt.plot(E_values, n_values, label=f'T = {T} K', marker='o' if T == 2000 else None)

# 
plt.xlabel('Energy E (eV)')
plt.ylabel('n(E, T)')
plt.title('Bose - Einstein Distribution at Different Temperatures')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 
a = 1
b = 0.6
c = 1
# 
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
u, v = np.meshgrid(u, v)
# 
x = a * np.sin(v) * np.cos(u)
y = b * np.sin(v) * np.sin(u)
z = c * np.cos(v)

# 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 
ax.plot_surface(x, y, z, label='Ellipsoid')

# 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
# 
ax.set_title('Ellipsoid Plot')
# 
ax.legend()
# 
plt.show()
```
