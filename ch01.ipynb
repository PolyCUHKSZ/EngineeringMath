## The lambdify() function

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, sin, diff
from sympy.utilities.lambdify import lambdify

a, b, t = symbols('a, b, t', real=True)
f = a*sin(b*t)
deriv = diff(f,t)  # evaluate symbolic derivarive of f
print('v = dy/dt = ', deriv)
```

```python
tims = np.linspace(0, 6, 50)
y = lambdify(t, f.subs({a:1, b:2}))
plt.subplot(1, 2, 1)  # first subplot, positions y(t)
plt.plot(tims, y(tims))
plt.xlabel('Time [s]')
plt.ylabel('y(t)')

u = deriv.subs({a:1, b:2})
v = lambdify(t, u)

plt.subplot(1, 2, 2)
plt.plot(tims, v(tims))
plt.xlabel('Time [s]')
plt.ylabel('Speed v(t)=dy/dt')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 7. Defining and plotting piecewise functions
import numpy as np
import matplotlib.pyplot as plt

# 定义分段函数
def f(x):
    conditions = [x < 5.0, x >= 5.0]
    functions = [lambda x: x, lambda x: x**2]
    return np.piecewise(x, conditions, functions)

# 测试单个值
print(f(3), f(7))

# 生成数组并测试
x = np.linspace(0, 10, 100)
y = f(x)

# 打印结果
# print(y)

# 绘制函数图像（可选）
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Piecewise Function')
plt.grid(True)
plt.show()
    
```

```python
# 8. The fibonacci series
def fibonacci(n=17):
    if n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

fibonacci()
```

```python
def fibonacci_v2(n=17):
    """
    生成斐波那契数列的前 n 项。

    参数:
    n (int): 要生成的斐波那契数列的项数，默认为 17。

    返回:
    list: 包含斐波那契数列前 n 项的列表。
    """
    if n == 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    # 初始化斐波那契数列的前两项
    res = [0, 1]
    a, b = 0, 1
    # 循环生成剩余的项
    for _ in range(2, n):
        a, b = b, a + b
        res.append(b)
    return res


# 调用函数并打印结果
print(fibonacci_v2())
```

```python
# 13.Evaluating of product series for cosx
import numpy as np


def f(x, N):
    product = 1
    for k in range(1, N + 1):
        term = 1 - (4 * x ** 2) / (np.pi ** 2 * (2 * k - 1) ** 2)
        product *= term
    return product

# 真实的cos(1)的值
true_value = np.cos(1)

# 计算不同N值时的结果并比较
N_values = [10, 50, 100]
for N in N_values:
    result = f(1, N)
    print(f"N = {N} 时，计算结果: {result}")
    print(f"与 cos(1) 的差值: {np.abs(result - true_value)}")
    print("-" * 30)
```

```python
# 14. Evaluation of the sum of infinite series
from sympy import symbols, Sum, factorial, sin, cos, exp, log, pprint

# 定义符号变量
x, n, k = symbols('x n k')

# 计算第一个级数
series1 = Sum((-1)**n * x**(2*n + 1) / factorial(2*n + 1), (n, 0, float('inf')))
result1 = series1.doit()

# 计算第二个级数
series2 = Sum(x**(2*n) / factorial(2*n), (n, 0, float('inf')))
result2 = series2.doit()

# 计算第三个级数
series3 = Sum((-1)**n * x**(2*n) / factorial(2*n), (n, 0, float('inf')))
result3 = series3.doit()

# 计算第四个级数
series4 = Sum(x**k / factorial(k), (k, 0, float('inf')))
result4 = series4.doit()

# 计算第五个级数
series5 = Sum(x**k / k, (k, 1, float('inf')))
result5 = series5.doit()

# 打印结果
print("第一个级数的结果:")
pprint(result1)
print("\n第二个级数的结果:")
pprint(result2)
print("\n第三个级数的结果:")
pprint(result3)
print("\n第四个级数的结果:")
pprint(result4)
print("\n第五个级数的结果:")
pprint(result5)
```

```python
# 17. The Bessel fucntions
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# 定义符号变量
x, n, p = sp.symbols('x n p')
# 定义伽马函数
gamma = sp.gamma

# 定义贝塞尔函数的级数表达式
bessel_series = sp.Sum(((-1)**n / (gamma(n + 1) * gamma(n + 1 + p))) * (x / 2)**(2 * n + p), (n, 0, sp.oo))

# 定义计算贝塞尔函数的函数
def bessel_function(p_val, x_val):
    result = bessel_series.subs({p: p_val, x: x_val}).doit()
    return result

# 生成 x 的数值范围
x_vals = np.linspace(-20, 20, 300)

# 计算 p = 0 和 p = 2 时的贝塞尔函数值
y_vals_0 = [float(bessel_function(0, x)) for x in x_vals]
y_vals_2 = [float(bessel_function(2, x)) for x in x_vals]

# 绘图
plt.plot(x_vals, y_vals_0, label=r'$J_0(x)$')
plt.plot(x_vals, y_vals_2, label=r'$J_2(x)$')
plt.xlabel('x')
plt.ylabel('$J_p(x)$')
plt.title('Bessel Functions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 18. Plots of the sum of terms in a Fourier series
import numpy as np
import matplotlib.pyplot as plt

# 定义周期 L
L = 5

# 定义 x 的取值范围
x = np.linspace(0, 10, 1000)

# 定义傅里叶级数求和函数
def fourier_series_sum(N, x):
    sum_value = 0
    for n in range(1, N + 1):
        sum_value += ((-1)**(n - 1) / n) * np.sin((n * np.pi / L) * x)
    return (50 / np.pi) * sum_value - 25

# 分别计算 N = 5, 10, 20, 30 时的傅里叶级数和
N_values = [5, 10, 20, 30, 100]
for N in N_values:
    y = fourier_series_sum(N, x)
    plt.plot(x, y, label=f'N = {N}')

# 设置图像属性
plt.xlabel('x')
plt.ylabel('T(x)')
plt.title('Fourier Series of Sawtooth Wave')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 19. Plots of the sum of terms in a Taylor series
import numpy as np
import matplotlib.pyplot as plt
import math


# 定义泰勒级数求和函数
def taylor_series_sum(N, x):
    sum_value = 0
    for n in range(N):
        term = ((-1) ** n / math.factorial(2 * n+1)) * (x ** (2 * n+1))
        sum_value += term
    return sum_value


# 生成x的取值范围
x = np.linspace(0, 8, 1000)

# 分别计算N = 5, 6, 8, 10时的泰勒级数和
N_values = [5, 6, 8, 10, 40]
for N in N_values:
    y = [taylor_series_sum(N, xi) for xi in x]
    plt.plot(x, y, label=f'N = {N}')

# 设置图像属性
plt.xlabel('x')
plt.ylabel('T(x)')
plt.title('Taylor Series of cos(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 21. The Pauli matrics
import numpy as np

# 定义泡利矩阵
sigma_x = np.array([[0, 1], [1, 0]])
sigma_y = np.array([[0, -1j], [1j, 0]])
sigma_z = np.array([[1, 0], [0, -1]])
identity_matrix = np.array([[1, 0], [0, 1]])

# 验证sigma_x^2 = I
result1 = np.dot(sigma_x, sigma_x)
assert np.allclose(result1, identity_matrix)

# 验证sigma_y^2 = I
result2 = np.dot(sigma_y, sigma_y)
assert np.allclose(result2, identity_matrix)

# 验证sigma_z^2 = I
result3 = np.dot(sigma_z, sigma_z)
assert np.allclose(result3, identity_matrix)

# 验证 -i * sigma_x * sigma_y * sigma_z = I
result4 = -1j * np.dot(np.dot(sigma_x, sigma_y), sigma_z)
assert np.allclose(result4, identity_matrix)

# 验证sigma_x * sigma_y = i * sigma_z
result5 = np.dot(sigma_x, sigma_y)
assert np.allclose(result5, 1j * sigma_z)

# 验证sigma_x * sigma_y + sigma_y * sigma_x = 0
result6 = np.dot(sigma_x, sigma_y) + np.dot(sigma_y, sigma_x)
assert np.allclose(result6, np.zeros((2, 2)))

# 验证sigma_x * sigma_y - sigma_y * sigma_x = 2i * sigma_z
result7 = np.dot(sigma_x, sigma_y) - np.dot(sigma_y, sigma_x)
assert np.allclose(result7, 2j * sigma_z)

print("所有恒等式验证通过！")
```

```python
from sympy import symbols, Matrix, cos, sin, trigsimp

# 定义符号
phi1, phi2 = symbols('phi1 phi2')

# 定义旋转矩阵B1和B2
B1 = Matrix([[cos(phi1), -sin(phi1)],
             [sin(phi1), cos(phi1)]])
B2 = Matrix([[cos(phi2), -sin(phi2)],
             [sin(phi2), cos(phi2)]])

# 计算两个矩阵的乘积
product_matrix = B1 * B2

# 计算单个旋转矩阵，旋转角度为phi1 + phi2
single_rotation_matrix = Matrix([[cos(phi1 + phi2), -sin(phi1 + phi2)],
                                 [sin(phi1 + phi2), cos(phi1 + phi2)]])

# 对乘积矩阵进行三角函数化简
simplified_product_matrix = product_matrix.applyfunc(trigsimp)

# 验证简化后的乘积矩阵与单个旋转矩阵是否相等
assert simplified_product_matrix == single_rotation_matrix

print("验证通过：两次旋转等同于一次旋转角度为phi1 + phi2的旋转。")
```

```python
# 24. Transpose and inverse of a matrix
import numpy as np

# 生成3x3的矩阵M，元素为0到2之间的随机整数
M = np.random.randint(0, 3, size=(3, 3))

# 计算矩阵M的转置
M_transpose = M.T

# 计算M与M转置的乘积
product = np.dot(M, M_transpose)

# 验证乘积是否为对称矩阵
is_symmetric = np.allclose(product, product.T)

print("矩阵M:")
print(M)
print("矩阵M的转置:")
print(M_transpose)
print("M与M转置的乘积:")
print(product)
if is_symmetric:
    print("M与M转置的乘积是对称矩阵。")
else:
    print("M与M转置的乘积不是对称矩阵。")
```

```python
from sympy import symbols, simplify, I

# 定义符号
R, L, C, omega = symbols('R L C omega')
Z_R = R
Z_L = I * omega * L
Z_C = -I / (omega * C)

# 计算总阻抗Z
Z = Z_R + Z_L + Z_C

# 先对Z进行化简
Z_simplified = simplify(Z)

# (a) 手动分离计算总阻抗Z的实部和虚部
real_part = simplify(Z_simplified.coeff(I, 0))
imag_part = simplify(Z_simplified.coeff(I, 1))

print("总阻抗Z的实部：", real_part)
print("总阻抗Z的虚部：", imag_part)

# (b) 代入数值计算Z的值
R_value = 5
L_value = 1e-3
C_value = 1e-3
omega_value = 10

Z_value = Z_simplified.subs({R: R_value, L: L_value, C: C_value, omega: omega_value})

print("代入数值后总阻抗Z的值：", Z_value)
```

```python
# 26. Finding maxima, minima and inflection points
from sympy import symbols, diff, solve, Eq, simplify

# 定义符号变量
x = symbols('x', positive=True)

# 定义函数
f = x**(-6) - x**(-12)

# 计算一阶导数
f_prime = diff(f, x)

# 计算二阶导数
f_double_prime = diff(f, x, 2)

# 求解驻点（一阶导数为0的点）
critical_points = solve(Eq(f_prime, 0), x)

# 求解拐点（二阶导数为0的点）
inflection_points = solve(Eq(f_double_prime, 0), x)

# 确定极大值/极小值点
max_min_points = []
for cp in critical_points:
    # 将驻点代入二阶导数判断是极大值还是极小值
    result = f_double_prime.subs(x, cp)
    if result > 0:
        max_min_points.append((cp, "min"))
    elif result < 0:
        max_min_points.append((cp, "max"))

print("一阶导数:")
print(f_prime)
print("二阶导数:")
print(f_double_prime)
print("极大值/极小值点:")
for point in max_min_points:
    print(f"x = {point[0]}, 类型: {point[1]}")
print("拐点:")
for ip in inflection_points:
    print(ip)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 玻尔兹曼常数，单位 J/K
k_B = 1.380649e-23
# 将电子伏特转换为焦耳
eV_to_J = 1.602176634e-19

# 定义Bose - Einstein分布函数
def bose_einstein_distribution(E, T):
    E_J = E * eV_to_J
    denominator = np.exp(E_J / (k_B * T)) - 1
    # 添加条件判断避免除零
    result = np.where(denominator != 0, 1 / denominator, np.nan)
    return result

# 能量范围，单位 eV
E_values = np.linspace(0, 30, 1000)

# 温度值，单位 K
T_values = [2000, 4000, 6000]

# 绘图
for T in T_values:
    n_values = bose_einstein_distribution(E_values, T)
    plt.plot(E_values, n_values, label=f'T = {T} K', marker='o' if T == 2000 else None)

# 设置图像属性
plt.xlabel('Energy E (eV)')
plt.ylabel('n(E, T)')
plt.title('Bose - Einstein Distribution at Different Temperatures')
plt.legend()
plt.grid(True)
plt.show()
```
